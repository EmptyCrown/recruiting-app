## GET GLASSDOOR DATA


```python
companies = []
extra = ["Jane Street", "Citadel", "Optiver", "Squarespace", "Wolverine Trading", "DRW", \
          "IMC Financial", "D.E.Shaw", "Old Mission Capital", "Tower Research", "Two Sigma", \
          "Hudson River Trading", "Ellington", "Town Square Trading", "Virtu Financial", "Vatic Labs", "Rubrik", \
         "Nauto", "Quora", "Snapchat", "Palantir", "Addepar"]
```

```python
for i in range(1000):
    # authentication information & other request parameters
    params_gd = OrderedDict({
        "v": "1",
        "format": "json",
        "t.p": "204455",
        "t.k": "cbcOhZMtUHK",
        "action": "employers",
        "employerID": "11111",
        # programmatically get the IP of the machine
        "userip": '65.112.10.26',
        "useragent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.81 Safari/537.36",
        "pn": i
    })

    # construct the URL from parameters
    basepath_gd = 'http://api.glassdoor.com/api/api.htm'

    # request the API
    response_gd = requests.get(basepath_gd,
                               params=params_gd,
                               headers={
                                   "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.81 Safari/537.36"
                               })
    # check the response code (should be 200)  & the content

    dat = json.loads(response_gd.content)
    companies += dat['response']['employers']
```
