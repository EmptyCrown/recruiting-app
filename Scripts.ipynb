## GET GLASSDOOR DATA


```python
companies = []
# extra = ["Jane Street", "Citadel", "Optiver", "Squarespace", "Wolverine Trading", "DRW", \
#           "IMC Financial", "D.E.Shaw Research", "D.E.Shaw Investment" "Old Mission Capital", "Tower Research", "Two Sigma", \
#           "Hudson River Trading", "Ellington Financial", "Town Square Trading", "Virtu Financial", "Vatic Labs", "Rubrik", \
#          "Nauto", "Quora, Inc.", "Snapchat", "Palantir Technologies", "Addepar", "Susquehanna International Group", "Bessemer Venture", "Sequoia Capital", "KPCB", "Accel Partners", \
#          "YCombinator", "8VC", "Asana", "Ctrl-labs", "Stripe"]
extra = ["Bessemer Ventures", "Jump Trading", "MongoDB", "Delphix", "Sumo Logic", "Thumbtack", "Docker", "Slack", "Robinhood", "Zenreach"]
```

```python
for q in extra:
    # authentication information & other request parameters
    params_gd = OrderedDict({
        "v": "1",
        "format": "json",
        "t.p": "204455",
        "t.k": "cbcOhZMtUHK",
        "action": "employers",
        "employerID": "11111",
        # programmatically get the IP of the machine
        "userip": '65.112.10.26',
        "useragent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.81 Safari/537.36",
        "q": q
    })

    # construct the URL from parameters
    basepath_gd = 'http://api.glassdoor.com/api/api.htm'

    # request the API
    response_gd = requests.get(basepath_gd,
                               params=params_gd,
                               headers={
                                   "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.81 Safari/537.36"
                               })
    # check the response code (should be 200)  & the content

    dat = json.loads(response_gd.content)
    companies += dat['response']['employers']
```

```python
C = {}
names = {}
```

```python
with open('js/companiesExtra.json') as data_file:    
    data = json.load(data_file)
    for company in data["companies"]:
        if not company["name"] in names:
            key = ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase + string.digits, k=12))
            C[key] = company
        names[company["name"]] = 1
        
```

```python
len(C.keys())
```

```python
with open('companiesFirebase.json', 'w') as outfile:
    json.dump(C, outfile)
```
